```python
import json

with open('pub_games.json', 'r') as file:
    data1 = json.load(file)

match_ids1 = sorted([match['match_id'] for match in data1])

max = match_ids1[-1]

data = [match_ids1[0]/max, 1.0]
```

```python
match_ids1[0]/max, match_ids1[-1]/max
```

```python
len(match_ids1), len(set(match_ids1))
```

```python
import matplotlib.pyplot as plt


plt.plot(range(len(data)), data, 'ro')  # 'ro' means red circles
plt.xticks(range(len(data)), range(len(data)))  # Set x-axis ticks to position numbers
plt.xlabel('Position')
plt.ylabel('Float Numbers')
plt.title('Plot of Float Numbers on 1D Axis with Positions')
plt.show()
```
