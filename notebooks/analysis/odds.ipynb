```python
import pandas as pd
import dotabet
import os
from datetime import datetime
features_csv_path = r"D:\WORKSPACE\dotabet\notebooks\parse_web\betting_odds.csv"

odds = pd.read_csv(features_csv_path)
```

```python
df = odds.dropna()
df = df[df['Match Date'] != 'LIVE']

def parse_date(d):
    try:
        d = d.replace('\r\n', ' ')
        return datetime.strptime(d, '%m/%d/%y %I:%M %p')
    except ValueError:
        return pd.NaT  # pd.NaT represents Not a Time (similar to NaN)

df['Match Date'] = df['Match Date'].apply(parse_date)
df['Match Date'] = df['Match Date'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M') if pd.notna(x) else x)
```
